```python
import altair as alt
import pandas as pd
import requests
import os
from io import StringIO
import matplotlib as plt
```

```python
url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
response = requests.get(url)
df = pd.read_csv(StringIO(response.text))
df_sampled = df.sample(n=500)


bar = alt.Chart(df_sampled).mark_bar(cornerRadius=10).encode(
    alt.X("Square Footage:Q", bin=True, title="Square Footage"),
    alt.Y("sum(Square Footage):Q", title="The Square Footage"),
    color=alt.value("red")  
).properties(
    title="The Distribution of Types Properties by Square Footage"
)

bar.display()

selection = alt.selection_interval()
```

```python
scatter= alt.Chart(df_sampled).mark_point().encode(
    x=alt.X('Square Footage:Q', title="The Square Footage"),
    y=alt.Y('Square Footage:Q', title="The Square Footage"),
    color=alt.condition(selection, alt.value("red"), alt.value("grey"))
).add_selection(
    selection
).properties(
    title="Scatter Plot of the Square Footage : Interactive"
)

scatter.display()
```

```python
directory = 'Visualizations/'
if not os.path.exists(directory):
    os.makedirs(directory)

bar.save(directory + 'bar_chart_final.html')

scatter.save(directory + 'scatter_plot_final.html')
```
